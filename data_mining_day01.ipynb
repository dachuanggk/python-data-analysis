```python

import random 
# zhfont=matplotlib.font_manager.FontProperties(fname="/Library/Fonts/Songti.ttc")
# 11-12点每分钟的温度变化
x = range(61)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]
plt.figure(figsize=(15, 6),dpi=50)
plt.plot(x,y_shanghai,color="r",linestyle="--",label="上海")
plt.plot(x,y_beijing,color="b",linestyle="-.",label="北京")
plt.legend(loc="best")
# 添加刻度
# x 的刻度说明
x_label = ["11点{}分".format(i) for i in x]
plt.xticks(x[::5],x_label[::5])
plt.yticks(range(0, 40, 5))
plt.grid(True, linestyle="--", alpha=0.5)
plt.xlabel("time")
plt.ylabel("temp")
plt.title("时间变化")
plt.show()
```

```python
import random 
# zhfont=matplotlib.font_manager.FontProperties(fname="/Library/Fonts/Songti.ttc")
# 11-12点每分钟的温度变化
x = range(61)
y_shanghai = [random.uniform(15,18) for i in x]
y_beijing = [random.uniform(1,3) for i in x]
# 创建画布
# plt.figure(figsize=(15, 6),dpi=50)
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6), dpi=80)
# 绘图
axes[0].plot(x,y_shanghai,color="r",linestyle="--",label="上海")
axes[1].plot(x,y_beijing,color="b",linestyle="-.",label="北京")

axes[0].legend(loc="best")
axes[1].legend(loc="best")

# 添加刻度
# x 的刻度说明
x_label = ["11点{}分".format(i) for i in x]
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_label[::5])
axes[0].set_yticks(range(0, 40, 5))
# axes[1].set_xticks(x[::5],x_label[::5])
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_label[::5])
axes[1].set_yticks(range(0, 40, 5))

# 网格
axes[0].grid(True, linestyle="--", alpha=0.5)
axes[1].grid(True, linestyle="-.", alpha=0.5)
axes[0].set_xlabel("time")
axes[0].set_ylabel("temp")
axes[0].set_title("时间变化")
axes[1].set_xlabel("time")
axes[1].set_ylabel("temp")
axes[1].set_title("时间变化")
plt.show()
```
