```python
import matplotlib.pyplot as plt
# 使用魔发指令 可以省略 plt.show()
%matplotlib inline
```

```python
cat = plt.imread('cat.jpeg')
plt.imshow(cat)
```

```python
# 垂直方向颠倒
cat1 = cat[::-1]
plt.imshow(cat1)
```

```python
# 水平 垂直方向颠倒
cat2 = cat[::-1,::-1]
plt.imshow(cat2)
```

```python
cat3 = cat[::,::,::-1]
plt.imshow(cat3)
```

```python
cat4 = cat[::5,::5]
plt.imshow(cat4)
```
