```python
# concatenating  合并多个df
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['A','B','C','D'])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['A','B','C','D'])
df3 = pd.DataFrame(np.ones((3,4))*2,columns=['A','B','C','D'])
print(df1)
print(df2)
print(df3)
```

```python
res = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
print(res)
```

```python
# join ,['inner','outer']
df1 = pd.DataFrame(np.ones((3,4))*0,columns=['A','B','C','D'],index=[1,2,3])
df2 = pd.DataFrame(np.ones((3,4))*1,columns=['B','C','D','E'],index=[2,3,4])
print(df1)
print(df2)
```

```python
res = pd.concat([df1,df2],join='outer')
res
```

```python
res = pd.concat([df1,df2],join='inner',ignore_index=True)
res
```

```python
#join axes 按某个矩阵的index合并
res = pd.concat([df1,df2],axis=1,join_axes=[df1.index])
res
```

```python
# 合并多个用列表
res = df1.append(df2,ignore_index=True)
res
```

```python
s1 = pd.Series([1,2,3,4],index=['A','B','C','D'])
s1
```

```python
df1.append(s1,ignore_index=True)
```

```python
#how=['inner','outer','left','right']
res = pd.merge(df1,df2,on=['B','C','D'],how='outer')
res
```

```python
res = pd.merge(df1,df2,left_index=True,right_index=True,how='outer')
res
```

```python
# handle overlapping
boys = pd.DataFrame({'k':['K0','K1','K2'],'age':[1,2,3]})
girls = pd.DataFrame({'k':['K0','K0','K3'],'age':[4,5,6]})
res = pd.merge(boys,girls,on='k',suffixes=['_boys','_girls'],how='outer')
print(boys)
print(girls)
res
```

```python
# Series
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
data = data.cumsum()
data.plot()
plt.show()
```

```python
#DataFrame
data = pd.DataFrame(np.random.randn(1000,4),index=np.arange(1000),columns=['A','B','C','D'])
# print(data.head(10))
data = data.cumsum()
# plot methods:
# 'bar','hist','box','kde','area','scatter',...
data.plot()
plt.show()
```
