```python
import pandas as pd
import numpy as np
```

```python
stock_change = np.random.normal(0, 1, (10,5))
pd.DataFrame(stock_change)
```

```python
# 添加行索引
stock_code = ["股票{}".format(i) for i in range(stock_change.shape[0])]
pd.DataFrame(stock_change, index=stock_code)
```

```python
# 添加列索引
date = pd.date_range(start="20210101", periods=5, freq="B")
date
```

```python
data = pd.DataFrame(stock_change, index=stock_code, columns=date)
```
